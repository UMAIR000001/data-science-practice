```python
# 1. Library Book System
# Design a class named Book to manage books in a library.
# Each book should have instance variables: title, author, is_available.
# Use a class variable total_books to track the total number of books.
# Implement methods to borrow and return books.

class Book:
    total_books = 0

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True
        Book.total_books += 1

    def borrow(self):
        if not self.is_available:
            raise Exception("The book is already borrowed.")
        self.is_available = False
        print(f"Book borrowed successfully | Title: '{self.title}' | Author: {self.author} |\n")

    def return_book(self):
        self.is_available = True

# Test
book1 = Book("Think and Grow Rich", "Napoleon Hill")
book1.borrow()

book2 = Book("Simplicity", "Umair")
book2.borrow()

print(f"Total number of books in the library: {Book.total_books}")
```

```python
# 4. Bank Account System
# Create BankAccount class with owner and balance.
# Implement deposit, withdraw with error handling


class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.__balance=balance
    
    def withdraw(self,amount): # Withdrawing Function
        if 0<amount<self.__balance:
            self.__balance-=amount
            print(f"After the Withdraw of {amount} the current Balance is {self.__balance}")
        else:
            print(f"Insufficient balance to withdraw")
    
    def deposit(self,amount): # Depositing Function
        if amount>0:
            self.__balance+=amount
            print(f"After the Deposit of {amount} the current Balance is {self.__balance}")
        else:
            print(f"Insufficient Funds to deposit")  
    def get_balance(self):
        print(f"My Current Balance is {self.__balance}")
        
account_1=BankAccount("Umair",500)
account_1.withdraw(33)
account_1.deposit(500)

account_1.get_balance()
```

```python
# 7. Employee Payroll System
# Create Employee class with salary calculation including bonus

class Employee:
    def __init__(self,name,basic_salary):
        self.name=name
        self.basic_salary=basic_salary
    
    def calculate(self):
        self.net_salary=round(self.basic_salary*1.10,1) # bonus of 10%
        return self.net_salary
        
employee1=Employee("RDJ",45000)
print(f"The employee {employee1.name} is our best asset having a basic salary of {employee1.basic_salary} with 10% bonus {employee1.calculate()}")
```

```python
# 8. Inventory Management System
# Create Product class with stock management and error handling
class Product:
    def __init__(self,name,price ,stock):
        self.name=name
        self.price=price
        self.stock=stock
    def restock(self,amount):
        self.stock+=amount
    def sell(self,units):
        if units>self.stock:
            raise "Sorry ! we don't have much units"
        self.stock-=units 
        
    
candy=Product("candy",200,5)
candy.restock(5)
print(f"The updated stock of product: {candy.name} and price is {candy.price} avaliable units: {candy.stock}")
candy.sell(9)
print(f"The current stock are {candy.stock}")
```

```python

                
#  9. Smart Thermostat Controller
# Create Thermostat class with temperature setting and mode toggling.
#                  

class Thermostat:
    
    max_temperature=40
    
    def __init__(self):
        self.temperature=20
        self.mode="Cooling"
        
    
    def set_temperature(self,temp):
        
        if temp>Thermostat.max_temperature:
            
            raise "The temperature is exceeding the maximum limit "
        
        self.temperature=temp
        
        return self.temperature
        
    def toggling(self):
        
        if self.mode=="Cooling":
             
             self.mode="Heating"
             
             return self.mode
        else:
             
             self.mode="Cooling"
             
             return self.mode
t=Thermostat()

print(f"Updated temperature is {t.set_temperature(30)}")

print(f"The mode is {t.toggling()}")

```

```python
# 10. Parking Lot System
# Create ParkingSpot class with park and free spot operations.
class ParkingSpot:
    total_spots = 0
    def __init__(self, spot_id):
        self.spot_id = spot_id
        self.is_occupied = False
        ParkingSpot.total_spots += 1
    def park_vehicle(self):
        if self.is_occupied:
            raise Exception("Spot already occupied.")
        self.is_occupied = True
    def free_spot(self):
        self.is_occupied = False
# Test
if __name__ == "__main__":
    spot = ParkingSpot("P1")
    spot.park_vehicle()
    assert spot.is_occupied == True
    spot.free_spot()
    assert spot.is_occupied == False
    assert ParkingSpot.total_spots == 1
    print("Parking Lot System tests passed.")
```
