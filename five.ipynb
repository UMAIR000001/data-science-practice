```python


def calculate_average(scores):
    total = 0  # Initialize total
    for x in scores:
        total += x  # Add each score to the total
    return total / len(scores)  # Calculate and return the average
exam_scores = [85, 92, 78, 90, 88, 95, 82, 79, 87, 91]  
average_score = calculate_average(exam_scores)  # Call the function
print("The Average of Exam scores is:", average_score)
```

```python
#               ---------Day No 06----------
# Example No 1 of user-defined Function 
def grade_distribution(grades):
    for grade in grades:
        if(90<=grade<=100):
            print("You got A grade ")
        elif(80<=grade<90):
            print("You got B grade ")
        elif(70<=grade<80):
            print("You got C grade ")
        elif(60<=grade<70):
            print("You got D grade ")
        elif(50<=grade<60):
            print("You got E grade ")
        elif(0<=grade<50):
             print("Failed! ")
        else:
            print("Error")
    #return  OPTIONAL
grades=[91,81,71,61,51,41,31,21,11,98]
grade_distribution(grades)
```

```python
#Example No 2 of user-defined function with the help of importing random.The below program will generate a secrete number and the user can guess it 
import random
def Number_guessing_game():
    secret_number=random.randint(1,100)
    guess=0
    attempts=0
    print("THe secret Number is :",secret_number)
    while(guess!=secret_number):
        guess=int(input("Guess the Number :"))
        attempts=attempts+1
        if(guess==secret_number):
            print("Congratulation you guessed the number ")
            print(f"You guessed the number in {attempts} ")
        elif(guess<secret_number):
            print("Too Low number")
        elif(guess>secret_number):
            print("Too High number")
           
        else:
            print("Invalid data ")
    

Number_guessing_game()
```

```python
# Example No 3
"""You are tasked to develop a program to calculate the sum of square of first n natural numbers,where n is entered by user"""
def add_natural_number(num):# User_defined Function
    n=1
    add=0
    while(n<=num):
        add+=n**2
        n=n+1
    print(add)
natural_num=int(input("Enter the ending point natural  number :"))
add_natural_number(natural_num)
    
```
